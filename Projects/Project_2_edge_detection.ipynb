# Image Filtering and Edge Detection Using Sobel Filters
## Introduction
- This project explores the fundamentals of image processing, focusing on the application of Sobel filters for edge detection. Edge detection is a critical operation in computer vision, enabling the identification of boundaries within images. By applying Sobel filters, we aim to highlight significant transitions in intensity, which often correspond to edges in the image.

## The Concept of Edge Detection
- Edge detection involves identifying significant changes in color or intensity between adjacent pixels in an image. These changes often represent boundaries of objects, outlines, or other important features within the image. Detecting edges is a foundational step in numerous image processing and computer vision tasks, such as object detection, feature extraction, and image segmentation.

## Role of Sobel Filters
- Sobel filters are used to detect edges in images by emphasizing areas of high spatial gradient. The operation consists of two convolutional kernels: one for detecting changes in intensity in the horizontal direction (Sobel X) and another for the vertical direction (Sobel Y). By convolving these kernels with the image, we obtain the gradient components in each direction, which can then be combined to find the overall gradient magnitude at each pixel.

## Gradient Magnitude and Edge Detection
- The gradient magnitude of an image at each pixel provides a measure of the edge strength. It is calculated from the horizontal (Gx) and vertical (Gy) gradient components obtained from the Sobel filters:



- Pixels with higher gradient magnitudes are likely to be part of edges. By selecting a threshold, we can identify significant edges in the image.

## Convolution Operation in Linear Algebra
- The convolution operation, central to applying Sobel filters, involves flipping a kernel (a small matrix) and moving it across the image. At each position, the kernel's values are multiplied by the corresponding pixel values in the image, and the results are summed up to produce a single pixel in the output image. This operation is a fundamental concept in linear algebra, demonstrating how matrix operations can be applied to real-world data.

## Calculus and Sobel Filter
- The Sobel filter operation is deeply rooted in calculus, particularly in the concept of derivatives. By detecting changes in intensity, the Sobel filter effectively computes a discrete derivative of the image in specific directions. This is analogous to finding the rate of change of the image's intensity function, highlighting areas where this rate changes abruptly (i.e., edges).

## Analyzing Edge Strengths
- The histogram of edge strengths (gradient magnitudes) can reveal much about the distribution of edges within an image. Areas with a high concentration of strong edges might indicate detailed or textured regions, while smoother areas will have fewer, weaker edges. Analyzing this histogram allows us to understand the image's structure and can inform further processing steps, such as thresholding to isolate significant edges.

## Conclusion
- Through this project, we've explored the application of Sobel filters for edge detection in images, demonstrating the relevance of concepts from linear algebra and calculus in processing and understanding visual data. The techniques and principles discussed here form the basis for more advanced image analysis and computer vision tasks.

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, filters, color

# load an image using skimage.io and convert the image to grayscale to simplify edge detection
image = io.imread('./nhl-camp.png')
image_gray = color.rgb2gray(image)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')
plt.show()

# apply Sobel Filters to detect horizontal and vertical edges
edge_sobel_h = filters.sobel_h(image_gray)
edge_sobel_v = filters.sobel_v(image_gray)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(edge_sobel_h, cmap='gray')
plt.title('Horizontal Edges')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edge_sobel_v, cmap='gray')
plt.title('Vertical Edges')
plt.axis('off')

plt.show()

# calculate Gradient Magnitude as a measure of edge strength
gradient_magnitude = np.sqrt(edge_sobel_h**2 + edge_sobel_v**2)

plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Magnitude')
plt.axis('off')
plt.show()

# statistical Analysis of Edge Strengths
print(f"Mean edge strength: {np.mean(gradient_magnitude)}")
print(f"Variance of edge strength: {np.var(gradient_magnitude)}")

plt.hist(gradient_magnitude.ravel(), bins=50, color='blue', alpha=0.7)
plt.title('Histogram of Edge Strengths')
plt.xlabel('Edge Strength')
plt.ylabel('Frequency')
plt.show()
```
