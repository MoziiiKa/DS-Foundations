# Gaussian Distribution Analysis

## Introduction to Gaussian Distribution
The Gaussian distribution, also known as the normal distribution, is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. It has a bell-shaped curve and is defined by two parameters: the mean (μ) and the standard deviation (σ). The formula for the Gaussian distribution is:



## Significance of Mean and Variance
- **Mean (μ):** Represents the central tendency of the data, indicating where the center of the distribution is located on the number line.
- **Variance (σ^2):** Measures the spread of the data around the mean. A small variance indicates that the data points are close to the mean, while a large variance suggests that the data points are spread out over a wider range of values.

## Relationship Between Histogram, PDF, and Probability
- **Histogram:** Provides a visual representation of the distribution of a dataset, showing the frequency of data points within specific intervals. It helps in understanding the underlying distribution of the data.
- **Probability Density Function (PDF):** For continuous data, the PDF represents the probability of a variable taking on a specific value. The area under the curve of a PDF (between two points) represents the probability of the variable falling within that interval.
- **Relation to Probability:** The histogram approximates the PDF as the sample size increases. The area under segments of the histogram estimates the probability of the variable falling within those segments, analogous to the integral of the PDF over an interval.

## Covariance
Covariance is a measure of how two variables change together. If the variables tend to increase and decrease together, the covariance is positive. If one variable tends to increase when the other decreases, the covariance is negative. In the context of Gaussian distributions, analyzing the covariance between pairs of variables helps in understanding their linear relationship.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data with Gaussian distribution
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f" % (data.mean(), data.std())
plt.title(title)
plt.show()

mean = np.mean(data)
variance = np.var(data)
print(f"Mean: {mean}, Variance: {variance}")
```
